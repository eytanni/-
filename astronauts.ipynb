# Data Check

```python
df['Year'] = pd.to_datetime(df['Year'])
```

```python
# Category Type
df.dtypes
```

```python
# Null Values
df.isnull().sum().sort_values(ascending=False)
```

```python
df.plot(subplots=True, layout=(5,4), kind='box', figsize=(12,14), patch_artist=True)
plt.subplots_adjust(wspace=0.5);
```

```python
features = ['Year', 'Gender','Group','Status', 'Military Rank','Space Flights','Space Walks','Death Mission']
fig, axes = plt.subplots(4,2,figsize=(15,15))
plt.subplots_adjust(wspace=0.5,hspace=0.3)
axes = axes.flatten()
for i, att in enumerate(features):
    sns.histplot(x=att, data=df, ax=axes[i])
```

# Gender Analysis:

```python
gender = df['Gender'].value_counts()

print(gender)

plt.figure(figsize=(10,10))
plt.pie(gender,labels=gender.index,autopct='%1.1f%%')
plt.title("GENDER")
plt.show()
```

```python
plt.figure(figsize=(5,5))
count_year = df['Year'].value_counts()
sns.countplot(y=df['Year'],order=count_year.nlargest(50).index).set(xlabel='Year',ylabel='Frequency')

plt.figure(figsize=(5,5))
count_gender = df['Year'].value_counts()
sns.countplot(y=df["Year"],hue=df["Gender"],order=count_gender.nlargest(50).index)
```

# Status Analysis

```python
status = df['Status'].value_counts()

print(status)

plt.figure(figsize=(10,10))
plt.pie(status,labels=status.index,autopct='%1.1f%%')
plt.title("STATUS")
plt.show()
```

Are there age restrictions?
There are no age restrictions for the program. Astronaut candidates selected in the past have ranged between the ages of 26 and 46, with the average age being 34.

Management astronauts are those experienced crew members who, because of their experience, have been promoted to other positions within NASA that are critical to the management of NASA's programs, or those astronauts on special duty assignments, sabbaticals, or administrative leave making them unavailable for direct support to the Astronaut Office. Depending on the needs of NASA, the astronaut can remain eligible for future space flight assignment and can return to active status at some future date. The management  assignments are considered career development for those astronauts who become mission commanders or leaders of major mission activities.  There are currently 36 astronauts at NASA in management status.
